```python
import pandas as pd
star_wars = pd.read_csv("star_wars.csv", encoding="ISO-8859-1")
star_wars.head(10)
```

```python
star_wars = star_wars[pd.notnull(star_wars["RespondentID"])]
star_wars["RespondentID"].value_counts()
```

```python
pd.unique(star_wars["Have you seen any of the 6 films in the Star Wars franchise?"])
```

```python
pd.unique(star_wars["Do you consider yourself to be a fan of the Star Wars film franchise?"])
```

```python
yn = ['Yes','No','Nan']
yn_to_tf = {
    'Yes':True,
    'No':False
}

cleanQ1 = star_wars["Have you seen any of the 6 films in the Star Wars franchise?"].map(yn_to_tf)
cleanQ2 = star_wars["Do you consider yourself to be a fan of the Star Wars film franchise?"].map(yn_to_tf)
cleanQ1
```

```python
star_wars["Have you seen any of the 6 films in the Star Wars franchise?"] = cleanQ1
star_wars["Do you consider yourself to be a fan of the Star Wars film franchise?"] = cleanQ2
```

```python
import numpy as np
sw_titles = pd.unique(star_wars.iloc[:,3:9].values.ravel("K"))
sw_titles_yn = {
}
for i in range(7):
    sw_titles_yn[sw_titles[i]] = True
sw_titles_yn[np.NaN] = False
for column in star_wars[star_wars.columns[3:9]]:
    star_wars[column] = star_wars[column].map(sw_titles_yn)
```

```python
row_names_old = star_wars.columns[3:9]
for i in range(6):
    star_wars = star_wars.rename(columns={
        row_names_old[i]:"seen_" + str(i+1)
    })
star_wars.head(10)
```

```python
star_wars[star_wars.columns[9:15]] = star_wars[star_wars.columns[9:15]].astype(float)
```

```python
row_names_old = star_wars.columns[9:15]
for i in range(6):
    star_wars = star_wars.rename(columns={
        row_names_old[i]:"ranking_" + str(i+1)
    })
star_wars.head(10)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
sw_movie_means = star_wars[star_wars.columns[9:15]].mean()
bar_positions = np.arange(6) + 0.75
plt.bar(bar_positions, sw_movie_means)
```

```python
#cleaned the Star Wars dataset to prepare to analyse how each 
#movie is ranked by respondents. We can see that the original
#trilogy is generally ranked higher, with Episode V having the 
#highest rank.
```

```python
seen_sums = star_wars[star_wars.columns[3:9]].sum()
plt.bar(bar_positions, seen_sums)
```

```python
#The original trilogy has been seen the most, as it was released 
#20 years before the prequels. Episode V is the most seen and 
#highest ranked, while Episode III is the least seen and lowest
#ranked.
```

```python
males = star_wars[star_wars["Gender"] == "Male"]
females = star_wars[star_wars["Gender"] == "Female"]
men_means = males[males.columns[9:15]].mean()
plt.bar(bar_positions, men_means)
```

```python
women_means = females[females.columns[9:15]].mean()
plt.bar(bar_positions, women_means)
```

```python
#Men and women rank the movies in roughly the same way
```

```python
men_sums = males[males.columns[3:9]].sum()
plt.bar(bar_positions, men_sums)
```

```python
women_sums = females[females.columns[3:9]].sum()
plt.bar(bar_positions, women_sums)
```

```python
#Men and women have similar patterns for which movies have been
#most seen
```
